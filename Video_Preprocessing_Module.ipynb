### Step 1: Import Required Libraries
This step imports only the libraries needed for video preprocessing.
No processing logic is included here so setup stays clear and beginner-friendly.

```python
# Step 1: Import required libraries only
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

### Step 2: Define Configuration Variables
This step defines all configurable preprocessing parameters in one place.
These parameters control video sampling and output frame size for landmark detection.

```python
# Step 2: Central configuration variables
VIDEO_PATH = "sample_classroom_video.mp4"
FRAME_SAMPLING_RATE = 10  # Select every 10th frame
RESIZE_WIDTH = 640
RESIZE_HEIGHT = 480
```

### Step 3: Load and Validate Video
This step opens the input video and validates that it can be read correctly.
It also prints key metadata (FPS and total frame count) for quick verification.

```python
# Step 3: Load video and run basic validation checks
if not os.path.exists(VIDEO_PATH):
    raise FileNotFoundError(f"Video file not found: {VIDEO_PATH}")

cap = cv2.VideoCapture(VIDEO_PATH)
if not cap.isOpened():
    raise RuntimeError("Failed to open video file. Check path and format.")

fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"Video loaded: {VIDEO_PATH}")
print(f"FPS: {fps:.2f}")
print(f"Total frames: {total_frames}")
```

### Step 4: Initialize Storage Variables
This step prepares counters and storage lists used in later cells.
No frame extraction or preprocessing is done here.

```python
# Step 4: Prepare counters and containers
frame_counter = 0
sampled_frames = []
print("Storage variables initialized.")
```

### Step 5: Frame Extraction with Sampling
This step reads the video frame-by-frame and keeps only every Nth frame.
No resizing, color conversion, or enhancement is applied yet.

```python
# Step 5: Extract sampled frames only
while True:
    success, frame = cap.read()
    if not success:
        break

    frame_counter += 1
    # Keep only every Nth frame
    if frame_counter % FRAME_SAMPLING_RATE == 0:
        sampled_frames.append(frame)

print(f"Total sampled frames: {len(sampled_frames)}")
```

### Step 6: Resize Frames
This step converts sampled frames into a fixed resolution for consistent processing.
Only resizing is done in this block to keep the pipeline modular.

```python
# Step 6: Resize sampled frames to a common size
resized_frames = []
for frame in sampled_frames:
    resized = cv2.resize(frame, (RESIZE_WIDTH, RESIZE_HEIGHT))
    resized_frames.append(resized)

print(f"Resized frames: {len(resized_frames)}")
```

### Step 7: Convert to RGB
This step converts frames from BGR to RGB for MediaPipe Holistic compatibility.
MediaPipe requires color input for accurate landmark detection. RGB frames are essential for face/hand/pose mesh analysis.

```python
# Step 7: Convert resized BGR frames to RGB
rgb_frames = []
for frame in resized_frames:
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    rgb_frames.append(rgb)

print(f"RGB frames ready: {len(rgb_frames)}")
```

### Step 8: Display Sample Frames
This step visualizes one sample frame before and after RGB conversion.
It shows how preprocessing prepares frames for landmark detection while maintaining color information.

```python
# Step 8: Display sample frames
if len(sampled_frames) == 0:
    print("No sampled frames available to display.")
else:
    sample_index = 0
    original_bgr = sampled_frames[sample_index]
    original_rgb = cv2.cvtColor(original_bgr, cv2.COLOR_BGR2RGB)
    resized_rgb = rgb_frames[sample_index]

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(original_rgb)
    plt.title("Original Frame (RGB)")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(resized_rgb)
    plt.title(f"Resized Frame ({RESIZE_WIDTH}x{RESIZE_HEIGHT})")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```

### Step 9: Clean Exit
This final step releases video resources and closes OpenCV windows.
Keeping cleanup explicit helps avoid locked files or hanging resources.

```python
# Step 9: Clean resource release
cap.release()
cv2.destroyAllWindows()
print("Cleanup complete.")
```
